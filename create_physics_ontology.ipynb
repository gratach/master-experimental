```python
from rdflib import Graph, Namespace, URIRef, Literal
```

```python
from pathlib import Path
datapath = Path("../master-database-files/master-experimental/physics_ontology/")
```

```python
def writeOntoArrayAsOWLOntology(ontoArray, filePath = str(datapath / "cPhyOnto.ttl"), bindPrefixes = []):
    g = Graph()
    g.bind("owl", "http://www.w3.org/2002/07/owl#")
    g.bind("rdf", "http://www.w3.org/1999/02/22-rdf-syntax-ns#")
    g.bind("rdfs", "http://www.w3.org/2000/01/rdf-schema#")
    for prefix, iri in bindPrefixes:
        g.bind(prefix, iri)
    for elem in ontoArray:
        if elem["type"] == "class":
            g.add((URIRef(elem["name"]), URIRef("http://www.w3.org/1999/02/22-rdf-syntax-ns#type"), URIRef("http://www.w3.org/2002/07/owl#Class")))
        elif elem["type"] == "property":
            g.add((URIRef(elem["name"]), URIRef("http://www.w3.org/1999/02/22-rdf-syntax-ns#type"), URIRef("http://www.w3.org/2002/07/owl#ObjectProperty")))
            g.add((URIRef(elem["name"]), URIRef("http://www.w3.org/2000/01/rdf-schema#domain"), URIRef(elem["domain"])))
            g.add((URIRef(elem["name"]), URIRef("http://www.w3.org/2000/01/rdf-schema#range"), URIRef(elem["range"])))
    g.serialize(destination=filePath, format='turtle')
```

```python
writeOntoArrayAsOWLOntology(cPhysOnto, bindPrefixes=[("cPhyOnto", cPhysOntoURI)])
```
